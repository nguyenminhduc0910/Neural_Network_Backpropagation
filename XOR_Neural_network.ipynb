```python
import numpy as np
import pandas as pd
```

```python
#ham sigmoid
def sigmoid(x):
  return 1/(1 + np.exp(-x))

#dao ham ham sigmoid
def sigmoid_derivative(x):
  return x*(1-x)
```

```python
class NeuralNetwork:
  def __init__ (self, layers, alpha=0.1):
    #mo hinh layers
    self.layers = layers
    # learning rate
    self.alpha = alpha
    # tham so W, b
    self.W = []
    self.b = []
    # khoi tao tham so 
    for i in range(len(layers)-1):
      # random theo phan phoi chuan
      w_ = np.random.randn(layers[i], layers[i+1])
      b_ = np.zeros((layers[i+1],1))
      self.W.append(w_/layers[i])
      self.b.append(b_)
  # tom tat mo hinh
  def __repr__ (self):
    # format thay the cho {}
    return "Newral Network [{}]".format('-'.join(str(l) for l in self.layers))
  
  #train mo hinh voi du lieu
  def fit_partial(self, x, y):
    A = [x]
    out = A[-1]
    #feedforward
    for i in range(0, len(self.layers)-1):
      out = sigmoid(np.dot(out, self.W[i])+ (self.b[i].T))
      A.append(out)
    #backpropagation
    y = y.reshape(-1,1)
    dA = [-(y/A[-1] - (1-y)/(1-A[-1]))]
    dW =[]
    db =[]
    for i in reversed(range(0, len(self.layers)-1)):
      dw_ = np.dot(A[i].T, dA[-1]*sigmoid_derivative(A[i+1]))
      db_ = (np.sum(dA[-1]*sigmoid_derivative(A[i+1]), 0)).reshape(-1,1)
      dA_ = np.dot(dA[-1]*sigmoid_derivative(A[i+1]), self.W[i].T)
      dW.append(dw_)
      db.append(db_)
      dA.append(dA_)

    # dao nguoc dW, db
    dW = dW[::-1]
    db = db[::-1]

    #gradient descent
    for i in range(len(self.layers)-1):
      self.W[i] = self.W[i] - self.alpha*dW[i]
      self.b[i] = self.b[i] - self.alpha*db[i]
    
  def fit(self, X, y, epochs =20, verbose = 10):
    for epoch in range(epochs):
      self.fit_partial(X,y)
      if epoch % verbose ==0:
        loss = self.calculate_loss(X,y)
        print('Epoch {}, loss {}'.format(epoch, loss))
  #du doan
  def predict(self,X):
    for i in range(len(self.layers)-1):
      X = sigmoid(np.dot(X, self.W[i])+ self.b[i].T)
    return X

  # loss function
  def calculate_loss(self, X, y):
    y_predict = self.predict(X)
    y = y.reshape(-1,1)
    #return np.sum((y-y_predict)**2)/2 neu la bai toan du doan
    return -np.sum(y*np.log(y_predict)+ (1-y)*np.log(1-y_predict))
```

```python
X = np.array([[1,1], [1,0], [0,1], [0,0]]*100)
y = np.array([0,1,1,0]*100)
model = NeuralNetwork([2,2,1], 0.001)
model.fit(X,y, 10000, 10)

X_test = np.array([[1,1], [1,0], [0,1], [0,0]])
y_test = np.array([0,1,1,0])
print(model.predict(X_test))
print('loss: ',model.calculate_loss(X_test,y_test))
```
